# Credits Data

```python
credits = pd.read_csv("/Users/jordan_uyeki/Documents/ucla/year4/spring2019/Stats131/Project/the-movies-dataset/credits.csv")
credits.head()
```

```python
#getting movie ids of the ones that are in the top series
movies_id = movies_in_series.id.tolist()
movies_id = [str(i) for i in movies_id]
```

```python
#subsetting series based on top movie series ids
series_credits = credits[credits['id'].isin(movies_id)]
series_credits.shape
```

```python
series_credits.head()
```

```python
#subsetting crew to just directors
series_credits['director_info'] = np.nan
keyValList = ['Director']
for i in range(0, len(series_credits)):
    temp = series_credits.crew.iloc[i]
    temp = ast.literal_eval(temp) 
    directors = [d for d in temp if d['job'] ==  'Director']
    series_credits.director_info.iloc[i] = directors
```

```python
series_credits.head()
```

```python
series_credits['num_directors'] = 0
for i in range(0, len(series_credits)):
    series_credits.num_directors.iloc[i] = len(series_credits.director_info.iloc[i])
series_credits.head()
```

```python
#making array of directors for each movie 
series_credits['directors'] = 0
for i in range(0, len(series_credits)):
    temp = series_credits.director_info.iloc[i]
    names = [d['name'] for d in temp]
    series_credits.directors.iloc[i] = names
series_credits.head()
```
