Analyzing top 10 production companies (and their divisions) based on https://reelrundown.com/film-industry/Top-10-Movie-Production-Companies 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv("/Users/jordan_uyeki/Documents/ucla/year4/spring2019/Stats131/Project/the-movies-dataset/movies_metadata.csv")
movies.head()
```

```python
movies.columns
```

```python
movies.shape
```

```python
movies = movies.sort_values(by = 'revenue', ascending = False)
```

```python
prod_companies = movies[pd.notnull(movies['production_companies'])]
prod_companies = prod_companies.sort_values(by = 'revenue', ascending = False)
prod_companies['belongs_to_series'] = prod_companies.belongs_to_collection.isna()
```

##  #1 Warner Bros.

```python
warner_bros = prod_companies[prod_companies['production_companies'].str.contains("Warner Bros.|New Line Cinema|Castle Rock|Turner Entertainment|DC Films")]
warner_bros.head()
```

```python
warner_bros.shape
```

```python
warner_bros.belongs_to_series.value_counts()
```

```python
warner_bros = warner_bros.sort_values(by = 'revenue', ascending = False)
warner_bros = warner_bros.iloc[1:101,];
warner_bros.belongs_to_series.value_counts()
```

Out of all of the movies produced by the Warner Bros, 19.3% of the movies belonged to a series.  Out of the top 100 movies produced by the Warner Bros, 65% of the movies belonged to a series.
